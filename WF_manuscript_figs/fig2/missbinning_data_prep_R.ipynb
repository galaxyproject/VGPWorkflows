```r
library(tidyverse)
library(scales)
```

```r
## FUNCTIONS
get2copy <- function(hist) {
  hist2copy <- hist %>% 
    filter(Copies == "2") %>% 
    filter(kmer_multiplicity >= 20) %>%
    filter(kmer_multiplicity <= 150)
  return(hist2copy)
}
getreadonly <- function(hist) {
  histreadonly <- hist %>% 
    filter(Copies == "read-only") %>% 
    filter(kmer_multiplicity >= 20) %>%
    filter(kmer_multiplicity <= 150)
  return(histreadonly)
}
process_spectra_2copy <- function(hist, prefix) {
  spectra_all <- read_tsv(hist)
  spectra_2copy <- get2copy(spectra_all)
  spectra_2copy <- add_column(spectra_2copy, Assembly = c(rep(prefix, nrow(spectra_2copy))))
  return(spectra_2copy)
}
process_spectra_readonly <- function(hist, prefix) {
  spectra_all <- read_tsv(hist)
  spectra_readonly <- getreadonly(spectra_all)
  spectra_readonly <- add_column(spectra_readonly, Assembly = c(rep(prefix, nrow(spectra_readonly))))
  return(spectra_readonly)
}
```

```r
## PROCESSING HISTOGRAMS
hic_hap1_2copy <- process_spectra_2copy("./bTaeGut2.trim.HiC.10x.bTaeGut2.trim.HiC.hic.hap1.p_ctg.spectra-cn.hist", "HiC hap1")
hic_hap1_readonly <- process_spectra_readonly("./bTaeGut2.trim.HiC.10x.bTaeGut2.trim.HiC.hic.hap1.p_ctg.spectra-cn.hist", "HiC hap1")
hic_hap2_2copy <- process_spectra_2copy("./bTaeGut2.trim.HiC.10x.bTaeGut2.trim.HiC.hic.hap2.p_ctg.spectra-cn.hist", "HiC hap2")
hic_hap2_readonly <- process_spectra_readonly("./bTaeGut2.trim.HiC.10x.bTaeGut2.trim.HiC.hic.hap2.p_ctg.spectra-cn.hist", "HiC hap2")
trio_hap1_2copy <- process_spectra_2copy("./bTaeGut2.trim.trio.10x.bTaeGut2.trio.cutadapt.20211115.dip.hap1.p_ctg.spectra-cn.hist", "Trio hap1 (pat)")
trio_hap1_readonly <- process_spectra_readonly("./bTaeGut2.trim.trio.10x.bTaeGut2.trio.cutadapt.20211115.dip.hap1.p_ctg.spectra-cn.hist", "Trio hap1 (pat)")
trio_hap2_2copy <- process_spectra_2copy("./bTaeGut2.trim.trio.10x.bTaeGut2.trio.cutadapt.20211115.dip.hap2.p_ctg.spectra-cn.hist", "Trio hap2 (mat)")
trio_hap2_readonly <- process_spectra_readonly("./bTaeGut2.trim.trio.10x.bTaeGut2.trio.cutadapt.20211115.dip.hap2.p_ctg.spectra-cn.hist", "Trio hap2 (mat)")
solo_pri_2copy <- process_spectra_2copy("./bTaeGut2.solo.p1p2.10x.bTaeGut2.solo.p1.spectra-cn.hist", "HiFi-only (pri)")
solo_pri_readonly <- process_spectra_readonly("./bTaeGut2.solo.p1p2.10x.bTaeGut2.solo.p1.spectra-cn.hist", "HiFi-only (pri)")
solo_alt_2copy <- process_spectra_2copy("./bTaeGut2.solo.p1p2.10x.bTaeGut2.solo.p2.spectra-cn.hist", "HiFi-only (alt)")
solo_alt_readonly <- process_spectra_readonly("./bTaeGut2.solo.p1p2.10x.bTaeGut2.solo.p2.spectra-cn.hist", "HiFi-only (alt)")
```
